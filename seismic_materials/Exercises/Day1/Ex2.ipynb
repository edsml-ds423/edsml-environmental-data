```python
data = np.load('../data/synthetic_cmp.npz')
list(data.keys())
```

```python
def plot(cmp, noffsets, nsamples, dt):
    """
    Plots synthetic cmp gathers
    """
    cutoff = 0.1
    plt.imshow(cmp, 
               extent=[0.5, noffsets + 0.5, dt*nsamples, 0], 
               aspect='auto',
               cmap='Greys',
               vmin=-cutoff,
               vmax=cutoff, 
               interpolation='none')
    
    # # following is purely for visual purposes
    # trace_numbers = list(range(1, noffsets + 1))
    # plt.xticks(trace_numbers)
    plt.title('CMP')
    plt.xlabel('Trace number')
    plt.ylabel('Time (s)')
    plt.show()
```

```python
dt = data['dt']
offsets = data['offsets']  
cmp = data['CMP'] 
data_points_per_offset = cmp.shape[0] # number of datapoints on each trace
number_of_offsets = cmp.shape[1] # number of traces

time_axis = np.arange(data_points_per_offset)*dt # time axis (proportional to depth)
# seis.plot(cmp, number_of_offsets, data_points_per_offset,dt)
plot(cmp, number_of_offsets, data_points_per_offset,dt)
```

## 2.2. Calculating NMO for our CMP Example

First, we are going to repeat what we did during the lecture and conduct an NMO correction using known velocities. This is to remind you how the NMO correction is done, and also this example is slightly more complicated as we have 3 reflections now not one. Once we have done that we will then move on to trying to determine the NMO velocities for yourselves!

```python
# known velocities **FOR THE TWO **REFLECTIONS** IN THE CMP GATHER**, see later how these are obtained in reality
v1, t1 = 3800, 0.22  # this has a less offset than v2, t2
v2, t2 = 4800, 0.46

# create velocity profile for **THE CMP GATHER**. 
# This will be used for the correction, different layers require different velocities to be corrected well
v_nmo = v1 + ((v2 - v1)/(t2 - t1))*(time_axis - t1)

print(len(v_nmo))

v_nmo[:10]
```

```python
# lets interpolate points on that line 
def interpolate(m: float, C: float, X: np.array):
    return m*X + C

# lets interpolate between the two points and extend it. 
x1, y1 = 0.22, 3800
x2, y2 = 0.46, 4800

m = (y2 - y1) / (x2 - x1)
C = y1 - m*x1

vnmo_no_shift = interpolate(m=m, C=C, X=time_axis)
vnmo_no_shift[:10]
```

```python
assert np.allclose(v_nmo, vnmo_no_shift)
```

```python
fig, ax = plt.subplots()
cutoff = 0.1
obj = ax.imshow(cmp,
                extent=[0, number_of_offsets, time_axis[-1], time_axis[0]],   # why do we add the extent here?
                aspect='auto',
                cmap='Greys',
                vmin=-cutoff,
                vmax=cutoff, 
                interpolation='none')

ax.scatter([0, 0], [time_axis[0], time_axis[-1]], c='r', marker='o', s=80, label='shifted time axis')

ax.set_xlabel('offset')
ax.set_ylabel('Time / s (data points per offset x dt)')
```

```python
# plotting explicit straight line interpolation along with the given inrterpolation
# also plotting the y-intercept

fig, ax = plt.subplots()
ax.scatter(time_axis, v_nmo, marker='o', s=10, color='green')
ax.scatter(time_axis, vnmo_no_shift, marker='o', s=5, color='orange')
ax.scatter([0, t1, t2], [C, v1, v2], marker='x', color='black', s=60)
ax.axvline(x=0, linestyle='--', color='black', alpha=0.4)
ax.set_xlim(0, np.max(time_axis) + (0.1*np.max(time_axis) ))
```

```python
# Comments:
# > by shifting the time axis so that the y-intercept is v1 at t=0 we only interpolate velocities for the 2 reflections
```

```python
## SEE SEMBLANCE PLOT BELOW FOR ANALYSIS

# vnmo is an array of interpolated velocities between the two points v1 and v2. 
# (v1, t1) and (v2, t2) represent the best velocities for the nmo correction of the 2 reflections that we are interested in reflecting.
# we got (v1, t1) and (v2, t2) by testing a range of velocities in a suitable range and finding the one that accentuated the amplitude the most. 
# The largest accentuation is a proxy for the best correction as the summed amplitudes will be the most aligned (horizontal)
# These two (v1, t1) and (v2, t2) values are also those that don't effect the other data points in the trace. 
# Only the two amplitudes of the reflections that we are interested in.
```

```python
fig, ax = plt.subplots()
cutoff = 0.1
obj = ax.imshow(cmp,
                # extent=[horizontal_min,horizontal_max,vertical_min,vertical_max]
                extent=[0, number_of_offsets, T[-1], T[0]],   # why do we add the extent here?
                aspect='auto',
                cmap='Greys',
                vmin=-cutoff,
                vmax=cutoff, 
                interpolation='none')

ax.set_xlabel('offset')
ax.set_ylabel('Time / s (data points per offset x dt)')


ax.axhline(y=t1, color='g', alpha=0.5)
ax.axhline(y=t2, color='g', alpha=0.5)
ax.scatter([0, 0], [t1, t2], c='g', s=20, label='t1 and t2')

# adding in times on the depth/time axis
ax.scatter([0, 0], [time_axis[0], time_axis[-1]], c='r', marker='o', s=20, label='raw time axis')
ax.scatter([0, 0], [(time_axis - t1)[0]*-1, (time_axis - t1)[-1]], c='r', marker='o', s=20, label='mod time axis')
ax.axhline(y=(time_axis - t1)[0]*-1, color='r', linestyle='dashed')
ax.axhline(y=(time_axis - t1)[-1], color='r', linestyle='dashed')

ax.legend(loc='best')

fig.colorbar(obj, ax=ax)
```

```python
nmo = seis.nmo_correction(cmp=cmp,
                          dt=dt,
                          offsets=offsets,
                          velocities=v_nmo)
nmo.shape
# you obtain a nmo array that contains an amplitude correction for each depth i for each offset.
```

```python
seis.plot(nmo, number_of_offsets, data_points_per_offset, dt)
```

As we saw in class, the NMO correction results in a stretch at large offsets (large trace numbers). This effect is more significant for shallower reflections with a short travel time. The direct arrival (the first seismic signal here) is not a reflection so it makes no sense to NMO correct it - that is why it looks so odd. When we process seismic data we apply a process called 'muting' which removes direct arrivals (it basically cuts them out of the data- we only want the reflections). 

## 2.3 Stacking NMO corrected data

Once we have done the NMO correction and the reflections have been corrected to horizontal it is now time to add them together in a process called 'stacking'- see lecture 3. The aim of stacking is to improve the signal to noise ratio. The CMP gather above will collapse to just a single trace where the reflections have a time of $t_0$ after stacking. 

```python
plt.plot(np.sum(nmo, axis = 1), time_axis, label='Sum of NMO corrected shot gather')  # sum the nmo horizontally to get a new amplitude along the x-axis (so 1200 points)
plt.plot(cmp[:,0], time_axis, label='Single trace, offset:' + str(offsets [0]))
plt.title("Signal/noise ratio increase by stacking")
plt.ylabel('Time (s)')
plt.xlabel('Amplitude')
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

In this example we used a synthetic CMP without noise. CMP stacking can, however, help you remove noise from noisy data. Since noise is partially random it will not sum up nicely like the NMO corrected reflections, so its amplitude in the stacked data will be low compared to the reflections. Have a go at the bonus exercise to try working with real CMP data.

##  2.2. Conducting a semblance analysis to determine NMO velocities

In reality, in order to find the correct NMO velocities for a full seismic trace we would need to do something called 'semblance analysis'- see lecture 3. This involves computing and applying NMO corrections for a wide range of velocities from lower than expected for the study area to higher. A calculation is done to assess the success of the NMO correction. In our simple example we are going to stack the NMO corrected gathers and calculate the maximum amplitude. When we have computed the NMO velocities correctly for a particular reflector we will get the highest stacked amplitudes. Doing this for all reflections within the seismic trace produces a semblance plot from which a seismic data processor can, in most cases manually, select the correct velocities to use.

```python
num_velocities_to_test = 10
V_test = np.linspace(1000, 7000, num_velocities_to_test)  # V testing range: [minimum velocity, maximum velocity, number of intervals/steps]

print(len(V_test))

amplitudes = seis.semblance(vs=V_test,
                            cmp=cmp,
                            dt=dt,
                            offsets=offsets,
                            ts=time_axis)

print(amplitudes.shape)

fig, ax = plt.subplots()
obj = ax.imshow(X=np.swapaxes(amplitudes, 0, 1),
                extent=[np.min(V_test), np.max(V_test), np.max(time_axis), 0],
                aspect="auto",
                cmap='jet')


# adding high amplitudes
ax.axhline(y=0.22, color='black', alpha=0.4, linestyle='dashed')
ax.axhline(y=0.455, color='black', alpha=0.4, linestyle='dashed')
ax.axvline(x=3700, color='black', alpha=0.4, linestyle='dashed')
ax.axvline(x=4200, color='black', alpha=0.4, linestyle='dashed')
ax.scatter([3700, 4200], [0.22, 0.455], c='black', s=20, marker='x')

ax.set_title(f'Semblance plot testing {len(V_test)} velocities between {np.min(V_test)} m/s and {np.max(V_test)} m/s.', fontsize=8)
fig.colorbar(obj, ax=ax, shrink=1).set_label(label = 'Corrected CMP amplitudes summation', size=6)
```

```python
# lets interpolate points on that line 
def interpolate(m: float, C: float, X: np.array):
    return m*X + C
```

```python
# lets interpolate between the two points and extend it. 
x1, y1 = 0.22, 3700
x2, y2 = 0.455, 4200

# points given to us in notebook.
# x1, y1 = 0.22, 3600
# x2, y2 = 0.46, 4800

m = (y2 - y1) / (x2 - x1)
C = y1 - m*x1

interp_v = interpolate(m=m, C=C, X=time_axis)
interp_v
```

```python
num_velocities_to_test = 10
V_test = np.linspace(1000, 7000, num_velocities_to_test)  # V testing range: [minimum velocity, maximum velocity, number of intervals/steps]

print(len(V_test))

amplitudes = seis.semblance(vs=V_test,
                            cmp=cmp,
                            dt=dt,
                            offsets=offsets,
                            ts=time_axis)

print(amplitudes.shape)

fig, ax = plt.subplots()
obj = ax.imshow(X=np.swapaxes(amplitudes, 0, 1),
                extent=[np.min(V_test), np.max(V_test), np.max(time_axis), 0],
                aspect="auto",
                cmap='jet')


# adding high amplitudes
ax.axhline(y=0.22, color='black', alpha=0.4, linestyle='dashed')
ax.axhline(y=0.455, color='black', alpha=0.4, linestyle='dashed')
ax.axvline(x=3700, color='black', alpha=0.4, linestyle='dashed')
ax.axvline(x=4200, color='black', alpha=0.4, linestyle='dashed')

# interpolated velocities, the t=0 is the x axis if you rotate the plot!
ax.plot(interp_v, time_axis, 'r-')

ax.scatter([3700, 4200], [0.22, 0.455], c='black', s=20, marker='x', zorder=0)

ax.set_title(f'Semblance plot testing {len(V_test)} velocities between {np.min(V_test)} m/s and {np.max(V_test)} m/s.', fontsize=8)
fig.colorbar(obj, ax=ax, shrink=1).set_label(label = 'Corrected CMP amplitudes summation', size=6)
```

```python
## SEE SEMBLANCE PLOT BELOW FOR ANALYSIS

# vnmo is an array of interpolated velocities between the two points v1 and v2. 
# (v1, t1) and (v2, t2) represent the best velocities for the nmo correction of the 2 reflections that we are interested in reflecting.
# we got (v1, t1) and (v2, t2) by testing a range of velocities in a suitable range and finding the one that accentuated the amplitude the most. 
# The largest accentuation is a proxy for the best correction as the summed amplitudes will be the most aligned (horizontal)
# These two (v1, t1) and (v2, t2) values are also those that don't effect the other data points in the trace. 
# Only the two amplitudes of the reflections that we are interested in.
```

### Comments:

- The main idea is that using the two velocities that give us the best amplitude for the two reflections when corrected, we can 
interpolate along the velocity line and get all of the corresponding velocitiy correction for each of the times (at dt intervals).
- If you look at the straight line, all of the interpolated velocities will not be correcting the plot to have "fake" amplitudes (ignoring the direct wave). 

#### Testing out how the number of velocity tests effects the semblance plot.

```python
ncols = 3
fig, axs = plt.subplots(1, ncols=ncols, figsize=(10,10))
intervals = [5, 10, 25] #, 10, 20, 50]  # number of velocities to be tested (semblance plot horizontal resolution)
for c in range(ncols):
    V = np.linspace(1000, 7000, intervals[c]) # V testing range: [minimum velocity, maximum velocity, number of intervals/steps]
    
    amplitudes = seis.semblance(vs=V,
                                cmp=cmp,
                                dt=dt,
                                offsets=offsets,
                                ts=time_axis)

    obj = axs[c].imshow(X=np.swapaxes(amplitudes, 0, 1),
                        extent=[np.min(V), np.max(V), np.max(time_axis), 0],
                        aspect="auto",
                        cmap='jet')
    
    axs[c].set_title(f'Semblance plot testing {len(V)} velocities.', fontsize=8)
    axs[c].set_xlabel('Velocity (m/s)', fontsize=6)

    # adding high amplitudes
    axs[c].axhline(y=0.22, color='black', alpha=0.4, linestyle='dashed')
    axs[c].axhline(y=0.455, color='black', alpha=0.4, linestyle='dashed')
    axs[c].axvline(x=3700, color='black', alpha=0.4, linestyle='dashed')
    axs[c].axvline(x=4200, color='black', alpha=0.4, linestyle='dashed')

    # interpolated velocities 
    axs[c].plot(interp_v, time_axis, 'r-')
 
    axs[c].scatter([3700, 4200], [0.22, 0.455], c='black', s=20, marker='x')




    fig.colorbar(obj, ax=axs[c], shrink=0.33).set_label(label = 'Corrected CMP amplitudes summation', size=6)

    if c == 0:
        axs[c].set_ylabel('Time (s)', fontsize=6)
        
fig.show()
```

# (1) reflection @ v1 ~ 3700 m/s & t1 = 0.22s

# (2) reflection @ v2 ~ 4200 m/s, t2 = 0.455s

Because we only have 4 seismic events in our synthetic CMP gather we don't have that many 'bulls eyes' in our semblance diagram. The direct arrival is not a reflection, so there the semblance plot at the depth of the direct arrival (<0.1 s) is nonsense and we can ignore it. We only care about finding the velocity for the three reflections. The reflection with a $t_0$ of around 0.2 clearly shows up in the semblace plot with a velocity around 3600 m/s. The low amplitude reflection with a $t_0$ around 0.4 doesn't really show up at all (maybe the amplitude was too low due to it being a **multiple- ASK A GTA IF YOU ARE INTERESTED IN WHAT A MULTIPLE IS**) so we can't determine a velocity for it. The reflection with a $t_0$ around 0.46 s gives a very wide bulls eye and it is difficult to pick the correct velocity precisely - this is a real feature of semblance analysis. It is difficult to get precise estimates of velocity at depth. In practice, you would select the velocities for the reflections you are confident in and then apply a (linear) interpolation between them. This will be the velocity profile you use to NMO correct your CMP gather for all travel times.
